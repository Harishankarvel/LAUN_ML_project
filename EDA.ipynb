```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully.")

sns.set_style('darkgrid')

file_name = 'ODI.csv'

try:
    df = pd.read_csv(file_name)
    print(f"Dataset '{file_name}' loaded successfully.")
except FileNotFoundError:
    print(f"Error: '{file_name}' not found.")
    print("Please make sure you have uploaded the dataset to the Colab session storage.")
    exit()

print("\n--- First 5 Rows of the Raw Dataset ---")
print(df.head())

print("\n--- Initial Dataset Info ---")
df.info()

df['is_not_out'] = df['Runs'].apply(lambda x: 1 if '*' in str(x) else 0)
df['Runs'] = df['Runs'].replace({r'\*': ''}, regex=True).astype(int)

# Convert 'Year' column to numeric, coercing errors to NaN and drop the 'Total' row
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df = df.dropna(subset=['Year'])
df['Year'] = df['Year'].astype(int)


print("\n--- Data Cleaning and Preprocessing Complete ---")
print("New columns 'is_not_out' and 'Year' added.")
print("\n--- First 5 Rows of Cleaned Dataset ---")
print(df.head())

print("\n--- Generating Visualizations ---")

yearly_runs = df.groupby('Year')['Runs'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Runs', data=yearly_runs, marker='o', color='b', linewidth=2)
plt.title("Rohit Sharma's Total ODI Runs Per Year", fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Runs', fontsize=12)
plt.xticks(yearly_runs['Year'], rotation=45)
plt.grid(True)
plt.show()

print("\n--- EDA Project Execution Complete ---")
print("All visualizations have been generated and displayed.")
```
